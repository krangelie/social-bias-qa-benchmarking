```python
import pandas as pd
```

```python
df = pd.read_excel("../../Qualitative coding/MAXQDA24 benchmark-use-in-llm-eval - Codesystem.xlsx")
df.head()
```

```python
df = df.rename(columns={"Liste der Codes": "code list", "Unnamed: 1": "category", "Unnamed: 2": "benchmark", "Unnamed: 3": "subclass", "Memo": "memo", "HÃ¤ufigkeit": "frequency"})
df.head()
```

```python
df["category"] = df["category"].fillna("")
current_category = ""
for i, row in df.iterrows():
    if len(row["category"]) > 1:
       current_category = row["category"]
    elif len(current_category) > 1:
        df.loc[i, "category"] = current_category
df
```

```python
benchmarks_df = df.loc[~df["benchmark"].isna(), ["category", "benchmark", "frequency"]]
benchmarks_df["relative frequency"] = benchmarks_df["frequency"] / 42 * 100 # Coded 42 benchmarks in total
benchmarks_df = benchmarks_df.sort_values(by="frequency", ascending=False)
benchmarks_df.head()
```

```python
benchmarks_df[benchmarks_df["benchmark"] == "PIQA"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(rc={"figure.figsize":(5, 9)}) #width=3, #height=4
benchmarks_df_reduced = benchmarks_df.loc[benchmarks_df["frequency"] >= 5]
ax = sns.barplot(x=benchmarks_df_reduced["relative frequency"] , y=benchmarks_df_reduced["benchmark"])
ax.set_title('Usage Frequency [%] per Benchmark', fontdict={'size': 15})
```

```python
sns.set_theme(rc={"figure.figsize":(5, 7)}) #width=3, #height=4
qa_benchmarks_df = benchmarks_df.loc[(benchmarks_df["category"] == "question answering") & (benchmarks_df["frequency"] >= 3)]
ax = sns.barplot(x=qa_benchmarks_df["relative frequency"] , y=qa_benchmarks_df["benchmark"])
ax.set_title('Usage Frequency [%] per QA Benchmark', fontdict={'size': 15})
```
